```python
#exo 1 data visualisation
import seaborn as sns
import pandas as pd
import warnings
import matplotlib
import numpy
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('data/house_pricing.csv')
#data.describe()
#pas de variables qualitatives
#quantiatives : toutes
#ces variables sont de type float ou string
```

```python
data.head()
```

```python
sns.relplot(x='LotArea', y='SalePrice', data=data)
#les pris sont Ã©lÃ©vÃ©s et nombreux pour une petite superficie 
```

```python
filter_lot_area_and_price = data[data.LotArea<20000][data.SalePrice<500000]
sns.relplot(x='LotArea', y='SalePrice', data=filter_lot_area_and_price)
```

```python
sns.relplot(x='LotFrontage', y='LotArea', data=data)
#les facades de ltos faibles ont une faible superficie
```

```python
filter_lot_area_and_price = data[data.LotFrontage<200][data.LotArea<100000]
sns.relplot(x='LotArea', y='SalePrice', data=filter_lot_area_and_price)
```
